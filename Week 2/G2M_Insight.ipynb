```python
ax = sns.countplot(x="City", data = cab);
plt.xticks(rotation=60);
plt.rcParams["figure.figsize"] = (20,10)
ax.set_title("Number of Cabs in Cities");
```

```python
ax = sns.displot(data=cab,x="Price Charged",kind="kde",hue="City");
ax.fig.set_size_inches(18,12)
ax.fig.subplots_adjust(top=0.9)
ax.fig.suptitle("Price Charged Density by Cities");
```

Categorical Values of Cab_Data's Company

```python
print(cab["Company"].value_counts())
color = ["deeppink","yellow"]
sns.set_palette(sns.color_palette(color))
ax= sns.countplot(x="Company", data = cab);

plt.rcParams["figure.figsize"] = (12,6)
ax.set_title("Number of Cabs Company in Cities");
```

```python
color = ["deeppink","yellow"]
sns.set_palette(sns.color_palette(color))
sns.displot(data=cab,x="Price Charged",kind="kde",hue="Company");
```

```python
color = ["deeppink","yellow"]
sns.set_palette(sns.color_palette(color))
sns.displot(data=cab,x="Cost of Trip",kind="kde",hue="Company");
```

```python
ax = sns.FacetGrid(cab, col="Company", margin_titles=True, height=4)
ax.map(plt.scatter, "KM Travelled", "Price Charged", color="purple", edgecolor="white", s=60, lw=1)
ax.set(xlim=(0, 55), ylim=(0, 2250));
ax.fig.set_size_inches(12,5)
ax.fig.subplots_adjust(top=.8) 
ax.fig.suptitle("Association between KM Travelled and Price Charged to Pink and Yellow Cab");
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y='Population',x='City',data=city.sort_values('Population',ascending=False))
plt.xticks(rotation=60);
plt.title("Population(M) in the Cities")
```
